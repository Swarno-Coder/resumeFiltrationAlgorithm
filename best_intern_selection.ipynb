## Importing modules & the file containing the dataset into pandas DataFrame

```python
import pandas as pd

resume = pd.read_csv('Applications_for_Machine_Learning_internship_edited.xlsx - Sheet1.csv')
resume.head()
```

## Finding total number of null elements to work with for Data Cleaning

```python
resume.isnull().sum()
```

## Renaming the column names/ feature names for better credability

```python
resume.rename(columns={'Python (out of 3)':'Python', 
       'Machine Learning (out of 3)':'ML',
       'Natural Language Processing (NLP) (out of 3)':'NLP',
       'Deep Learning (out of 3)':'DL', 
       'Are you available for 3 months, starting immediately, for a full-time work from home internship? ':'availability'
        },inplace=True)
resume.info()
```

### Also dropping the 'Name' column because it is unwanted for modelling

```python
resume = resume.drop('Name',axis=1)
```

## Now checking the which applicant is better knowing the key fields for the job 
---
NOTE: column name can be changes for different job requirement


### Go through all the steps below untill you found only one applicant left, 
### `STOP there`

```python
columns_to_check = ['Python', 'ML', 'NLP', 'DL']

# Check if each row has all zeros in the selected columns
rows_with_zeros = resume[columns_to_check].eq(3).all(axis=1)

# Remove rows with all zeros in the selected columns
newdf = resume.loc[rows_with_zeros]
newdf.shape, resume.shape
```

### Checking sum of null elements once more

```python
newdf.isnull().sum()
```

```python
newdf['availability'].unique()
```

### Dropping applicant who is not available currently

```python
newdf=newdf[~newdf['availability'].str.contains('No')]
```

### Dropping applicant who doesnot mention the degree at all

```python
newdf = newdf.dropna(subset=['Degree'])
newdf.shape
```

Checking NULL again

```python
newdf.isnull().sum()
```

```python
newdf
```

### Dropping the applicant who has not provide any of the Educational Qualification

```python
columns_to_check = ['Performance_PG',
       'Performance_UG', 'Performance_12', 'Performance_10']

# Check if each row has all zeros in the selected columns
rows_with_zeros = newdf[columns_to_check].isnull().all(axis=1)

# Remove rows with all zeros in the selected columns
finaldf = newdf.loc[~rows_with_zeros]
finaldf.shape
```

```python
finaldf
```

### Dropping the applicant don't have provided any of the Educational Qualification

```python
columns_to_check = ['Performance_PG',
       'Performance_UG', 'Performance_12', 'Performance_10']

# Check if each row has all zeros in the selected columns
rows_with_zeros = finaldf[columns_to_check].isnull().any(axis=1)

# Remove rows with all zeros in the selected columns
finaldf = finaldf.loc[~rows_with_zeros]
finaldf
```

## Changing Performance Numbers to float and then sorting by required feature/column

```python
import warnings, numpy as np
warnings.filterwarnings('ignore')
finaldf['Performance_UG'] = finaldf['Performance_UG'].str.extract(r'(\d+\.\d+)').astype(float)
finaldf['Performance_PG'] = finaldf['Performance_PG'].str.extract(r'(\d+\.\d+)').astype(float)
finaldf['Performance_12'] = finaldf['Performance_12'].str.extract(r'(\d+\.\d+)').astype(float)
finaldf['Performance_10'] = finaldf['Performance_10'].str.extract(r'(\d+\.\d+)').astype(float)
finaldf
```

```python
columnName = 'Current Year Of Graduation'
sorted_df = finaldf.sort_values(by=columnName, ascending=False)
sorted_df
```

```python
sorted_df.index[0]
```

## Sorting them in case of Higher Degree

```python
sorted_df = finaldf.sort_values(by='Degree', ascending=False)
sorted_df.index[0]
```

### Getting the details of the final selected applicant details for next round of interview

```python
sorted_df.iloc[[0]]
```

## Cleaning the dataframes for memory cleanup

```python
del newdf, finaldf, resume, sorted_df
```
